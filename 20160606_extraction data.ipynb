On peut donc maintenant filtrer les bonnes tables statistiques :

```python
def is_monthly(node):
    "Returns true if node is monthly data point."
    search =  node.find(class_='procedure_choix')
    if search is not None:
        return search.text == 'Ecoulement mensuel'
    else:
        return False
```

```python
stats = [node for node in soup.find_all(class_='statistiques') if is_monthly(node)]
```

Maintenant, il s'agit d'extraire les données utiles.

```python
def extract_height_date(node):
    "Returns height and date from node."
    p = node.find(class_='procedure_choix', text='Maximum instantané')
    try:
        h = p.parent.parent.parent.find('td', text='Hauteur :').parent.find(id='C3').text
        d = p.parent.parent.parent.find('td', text='Date :').parent.find(id='C3').text
        return (h, d)
    except AttributeError:
        return None
```

# A bit of pandas 

```python
import pandas as pd
```

On écrit une petite fonction qui fait un pipe :

```python
def value_to_float(df):
    "Converts Value to float."
    df = df.copy()
    df['Valeur'] = [float(val.replace(',', '.')) for val in df['Valeur']]
    return df
```

Et on ouvre le dataset :

```python
df = pd.read_csv(open('crues_seine.csv', encoding='iso-8859-1'), sep=';', parse_dates=['Date']).pipe(value_to_float).set_index('Date')
```

```python
df.groupby('Type').count()
```
